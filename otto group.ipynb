```python
from IPython.display import display, HTML
# print(y_train[:1])
train_df = pd.read_csv('train.csv')


# import labelencoder
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le = LabelEncoder()


# apply le on categorical feature columns
train_df['target'] = le.fit_transform(train_df['target'])


print(train_df['target'].value_counts())
display(train_df.head())

independant_cols = train_df.columns.tolist()
print(len(independant_cols))
independant_cols.remove('id')
independant_cols.remove('target')


X_train, X_test, y_train, y_test = train_test_split(train_df[independant_cols], train_df['target'], 
                                                    test_size=0.2, random_state=42)




from sklearn.metrics import classification_report,confusion_matrix,roc_curve,auc
dtree = DecisionTreeClassifier(random_state=0, max_depth=5  )
dtree.fit(X_train,y_train)


Y_train_predict = dtree.predict(X_train)
print(Y_train_predict[:5])
print(confusion_matrix(y_train,Y_train_predict))
print(classification_report(y_train,Y_train_predict))

Y_test_predict = dtree.predict(X_test)
print(confusion_matrix(y_test,Y_test_predict))
print(classification_report(y_test,Y_test_predict))
```
